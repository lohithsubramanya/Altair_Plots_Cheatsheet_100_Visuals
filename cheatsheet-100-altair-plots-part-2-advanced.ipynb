**Properties:**

* `Chart` - takes a dataframe as a single argument. alt.Chart(dataframe)
* `encode()`- builds a key-value mapping between encoding channels (such as x, y, color, shape, size, etc.) 

* `color`- The color of the mark
* `fill` - The fill for the mark
* `fillOpacity` - The opacity of the markâ€™s fill
* `opacity` - The opacity of the mark
* `shape` - The shape of the mark
* `size` - The size of the mark
* `stroke` - The stroke of the mark
* `strokeDash` - The stroke dash style
* `strokeOpacity` - The opacity of the line
* `strokeWidth` - The width of the line
* `text` - Text to use for the mark
* `tooltip` - The tooltip value
* `href` - Hyperlink for points
* `order` -	Sets the order of the marks

**Encoding Data Types:**

* **quantitative** - `Q`	- a continuous real-valued quantity
* **ordinal** -	`O` - a discrete ordered quantity
* **nominal** -	`N` -	a discrete unordered category
* **temporal** - `T` -	a time or date value
* **geojson** -	`G` -	a geographic shape

More attribute details here :https://altair-viz.github.io/user_guide/encoding.html

```python
base = alt.Chart(last_5_netflix).transform_aggregate(
    num_shows='count()',
    groupby=['type', 'release_year']
).encode(
    alt.X('type:O', scale=alt.Scale(paddingInner=0),title="Type of Program"),
    alt.Y('release_year:O', scale=alt.Scale(paddingInner=0),title="Released Year"),
).properties(width=580,height=400,title="Degree and Test Percentage of Placed Comm&mgmnt candidates")

# Configure heatmap
heatmap = base.mark_rect().encode(
    color=alt.Color('num_shows:Q',
        scale=alt.Scale(scheme='darkgold'),
        legend=alt.Legend(direction='horizontal')
    )
)

# Configure text
text = base.mark_text(baseline='middle').encode(
    text='num_shows:Q',
    color=alt.condition(
        alt.datum.num_shows > 500,
        alt.value('black'),
        alt.value('white')
    )
)

# Draw the chart
heatmap + text
```

<font color="green" size=+2.5><b>77. Simple Scatter Error Plot</b></font>

<a href="#top" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover" title="go to Colors">Go to TOC</a>

```python
educat=university[(university['quality_of_education']<6)][['quality_of_education','score']].groupby('quality_of_education').agg('mean').reset_index()
```

```python
yerr=4
source = pd.DataFrame({"x":educat['quality_of_education'], "y":educat['score'], "yerr":yerr})
base = alt.Chart(source).transform_calculate(
    ymin="datum.y-datum.yerr",
    ymax="datum.y+datum.yerr"
)
base

# generate the points
points = base.mark_point(
    filled=True,
    size=50,
    color='black'
).encode(
    x=alt.X('x',title="Quality of Education",scale=alt.Scale(domain=(0, 6))), 
    y=alt.Y('y', title="Score",scale=alt.Scale(domain=(70, 100))),
    tooltip=[alt.Tooltip('x',title="Quality of Education"),alt.Tooltip('y',title="Score")]
).properties(width=600,height=400,title="Quality of Education vs Score with Error bar of length 4")

#generate the error bars
errorbars = base.mark_errorbar().encode(
    x="x",
    y="ymin:Q",
    y2="ymax:Q"
)

points + errorbars
```

<font color="teal" size=+2.5><b>87. Ridgeline</b></font>

<a href="#top" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover" title="go to Colors">Go to TOC</a>

```python
weather = data.seattle_weather.url
```

```python
step = 20
overlap = 1

alt.Chart(weather, height=step).transform_timeunit(
    Month='month(date)'
).transform_joinaggregate(
    mean_temp='mean(temp_max)', groupby=['Month']
).transform_bin(
    ['bin_max', 'bin_min'], 'temp_max'
).transform_aggregate(
    value='count()', groupby=['Month', 'mean_temp', 'bin_min', 'bin_max']
).transform_impute(
    impute='value', groupby=['Month', 'mean_temp'], key='bin_min', value=0
).mark_area(
    interpolate='monotone',
    fillOpacity=0.8,
    stroke='lightgray',
    strokeWidth=0.5
).encode(
    alt.X('bin_min:Q', bin='binned', title='Maximum Daily Temperature (C)'),
    alt.Y(
        'value:Q',
        scale=alt.Scale(range=[step, -step * overlap]),
        axis=None
    ),
    alt.Fill(
        'mean_temp:Q',
        legend=None,
        scale=alt.Scale(domain=[30, 5], scheme='redyellowblue')
    )
).facet(
    row=alt.Row(
        'Month:T',
        title=None,
        header=alt.Header(labelAngle=0, labelAlign='right', format='%B')
    )
).properties(
    title='Seattle Weather',
    bounds='flush'
).configure_facet(
    spacing=0
).configure_view(
    stroke=None
).configure_title(
    anchor='end'
)
```

<a href="#top" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover" title="go to Colors">Go to TOC</a>


```python
def find_age_stay(start,end):
    result=np.random.randint(start,end)
    return result

health['Stay']=np.where(health['Stay']=="0-10","1-10",health['Stay'])
health['Stay']=np.where(health['Stay']=="More than 100 Days","101-110",health['Stay'])

health[['start_age','end_age']] = health['Age'].str.split('-',expand=True)
health[['start_stay','end_stay']] = health['Stay'].str.split('-',expand=True)
health['end_age']=health['end_age'].astype(int)+1
health['end_stay']=health['end_stay'].astype(int)+1

health['Age_num']=health[['start_age','end_age']].apply(lambda x:find_age_stay(*x), axis=1)
health['Stay_days']=health[['start_stay','end_stay']].apply(lambda x:find_age_stay(*x), axis=1)
health['Age_num']=health['Age_num'].astype(int)
health['Stay_days']=health['Stay_days'].astype(int)
age_stay=health[['Age_num','Stay_days']]
```

```python
selector = alt.selection_single(empty='all', fields=['Severity of Illness'])

color_scale = alt.Scale(domain=['Minor', 'Moderate','Extreme'],
                        range=['green', 'orange','red'])

base = alt.Chart(health.sample(5000)).properties(
    width=280,
    height=200
).add_selection(selector)

points = base.mark_point(filled=True, size=200).encode(
    x=alt.X('mean(Age_num):Q',title="Age - Mean",
            scale=alt.Scale(domain=[20,100])),
    y=alt.Y('mean(Stay_days):Q',title="Stay Days - Mean",
            scale=alt.Scale(domain=[0,60])),
    color=alt.condition(selector,
                        'Severity of Illness:N',
                        alt.value('lightgray'),
                        scale=color_scale),
)

hists = base.mark_bar(opacity=0.5, thickness=100).encode(
    x=alt.X('Age_num',
            bin=alt.Bin(step=5), # step keeps bin size the same
            scale=alt.Scale(domain=[0,100])),
    y=alt.Y('count()',
            stack=None,
            scale=alt.Scale(domain=[0,350])),

    color=alt.Color('Severity of Illness:N',
                    scale=color_scale)
).transform_filter(
    selector
)


points | hists
```
